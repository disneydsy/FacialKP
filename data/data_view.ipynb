```python
import pickle
pkl_file = open('data.pkl', 'rb')
data = pickle.load(pkl_file)
pkl_file.close()
train_data=data['train_data']
train_label=data['train_label']
test_data=data['test_data']
output_txt=data['output_txt']
print train_data.shape,train_label.shape,test_data.shape,output_txt
```

```python
from pylab import *
%matplotlib inline
plt.rcParams['figure.figsize'] = (15, 15)        # large images

for i in range(20):
    plt.subplot(4, 5, i+1)
    plt.imshow(train_data[7048-i].reshape((96,96)),cmap = plt.get_cmap('gray'))
    plt.scatter(x=train_label[7048-i][::2], y=train_label[7048-i][1::2])        
plt.show()
```

```python
for i in range(20):
    plt.subplot(4, 5, i+1)
    plt.imshow(test_data[i].reshape((96,96)),cmap = plt.get_cmap('gray'))     
plt.show()
```
